```python
## taking three data frames df1 ,df2 and df3
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3'],
                        'C': ['C0', 'C1', 'C2', 'C3'],
                        'D': ['D0', 'D1', 'D2', 'D3']},
                        index=[0, 1, 2, 3])
print(df1,"\n\n")
df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                        'B': ['B4', 'B5', 'B6', 'B7'],
                        'C': ['C4', 'C5', 'C6', 'C7'],
                        'D': ['D4', 'D5', 'D6', 'D7']},
                         index=[4, 5, 6, 7])
print(df2,"\n\n")
df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                        'B': ['B8', 'B9', 'B10', 'B11'],
                        'C': ['C8', 'C9', 'C10', 'C11'],
                        'D': ['D8', 'D9', 'D10', 'D11']},
                        index=[8,9,10,11])
print(df3,"\n\n")

```

```python
# concatenation of these above three  as df1 then df2 then df3
concat_data=pd.concat([df1,df2,df3])
print(concat_data)
concat_data.shape
#here you can see below the data is concatenated row wise means the df2 has been appended at the end of df1 similarly df3 \
#did . here under same column and different indexing
```

```python
## taking three data frames df1 ,df2 and df3  but one has same index   and different column  
# in column wise : same or different column  will be added and values will be filled based on their index

df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3'],
                        'C': ['C0', 'C1', 'C2', 'C3'],
                        'D': ['D0', 'D1', 'D2', 'D3']},
                        index=[0, 1, 2, 3])
print(df1,"\n\n")
df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                        'B': ['B4', 'B5', 'B6', 'B7'],
                        'C': ['C4', 'C5', 'C6', 'C7'],
                        'D': ['D4', 'D5', 'D6', 'D7']},
                         index=[4, 5, 6, 7])
print(df2,"\n\n")
df3 = pd.DataFrame({'v': ['A8', 'A9', 'A10', 'A11'],
                        'x': ['B8', 'B9', 'B10', 'B11'],
                        'y': ['C8', 'C9', 'C10', 'C11'],
                        'z': ['D8', 'D9', 'D10', 'D11']},
                        index=[0,1,2,3])
print(df3,"\n\n")
data=pd.concat([df1,df2,df3])  # here data has been added rows wise but under its own different column\ 
                                 #name and shows nan vlaues where matching is not
data=pd.concat([df1,df2,df3],axis=1)
data
```
