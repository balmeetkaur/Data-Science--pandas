```python
import seaborn as sns
import numpy as np
data = sns.load_dataset('titanic')
data
```

## replace column vales with new value

```python
data["sex"]=data["sex"].replace({"male":"M"})   ## replace Male with M 
data
```

```python
data["sex"]=data["sex"].replace("female","F")
data
```

```python
data["who"]=data["who"].replace("man",0)    ## encoding
data["who"]=data["who"].replace("woman",1)
data["who"]=data["who"].replace("child",2)
data
```
