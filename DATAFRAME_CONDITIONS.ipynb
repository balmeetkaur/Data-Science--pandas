## Conditions on columns 

```python
# fetching indexes of null values of column name age
import seaborn as sns
import numpy as np
data = sns.load_dataset('titanic')

np.where(data["age"].isnull()==True)



```

```python
# fetching those data where gae has misiing values using iloc
missing = data.iloc[np.where(data["age"].isnull()==True)]
missing
```

```python
#data.iloc[np.where(data.age==35)]
data[data.age==35] 
```

```python
#fetching the data where age==35 and who is woman
data.iloc[np.where((data.age==35) & (data.who=="woman"))] #or data[(data.age==35.0 )& (data.who=="woman")]
```

```python
data[(data["age"]==35.0) & (data["fare"]== data["fare"].max())]
```

```python
data[(data["age"]>=30.0) & (data["fare"]== data["fare"].max())]
```

```python
#either or : |
data[(data["age"]>=30.0) | (data["fare"]== data["fare"].max())]
```
