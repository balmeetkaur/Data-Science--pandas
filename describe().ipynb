### describe () :gives statistical details of data

```python
import seaborn as sns
import numpy as np
data = sns.load_dataset('titanic')
data.describe()    ##getting statistical details of numeric columns
```

```python
data.describe().T
```

```python
data.describe(include="object")  ## getting statistical details of categorical data
```

```python
data.describe(include="all")   ## getting statistical details of all columns
```
